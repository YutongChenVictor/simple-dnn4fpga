```python
import numpy as np
import pandas as pd
import sys

# 定义常量和初始值
SOC_MAX = 0.95
SOC_MIN = 0.0

SOH_MAX = 0.99
SOH_MIN = 0.60

VOLTAGE_MAX = 4.2
VOLTAGE_MID = 1.7
VOLTAGE_MIN = 0

CURRENT_MAX = 2

SIZE = 180
SIZE_MAX = 180
SIZE_MIN = 170

TEMP_MAX = 32
TEMP_MIN = 26


def random_int(low, high):
    return np.random.randint(low, high + 1)

def random_float(divisor):
    return np.random.rand() / divisor


all_data = []  # 移到外层，避免在每个温度循环中重新初始化

# 循环温度范围
for temp in range(26, 32):

    for num in range(100):
        size = random_int(SIZE_MIN, SIZE_MAX)
        soc = np.zeros(SIZE)
        soh = np.zeros(SIZE)
        voltage = np.zeros(SIZE)
        current = np.zeros(SIZE)

        # SOC
        if random_int(0,1):
            soc[0] = SOC_MAX + random_float(80)
        else:
            soc[0] = SOC_MAX - random_float(80)

        # SOH
        if random_int(0,1):
            soh[0] = SOH_MAX + random_float(80)
        else:
            soh[0] = SOH_MAX - random_float(80)

        # Voltage
        if random_int(0,1):
            voltage[0] = VOLTAGE_MAX + random_float(80)
        else:
            voltage[0] = VOLTAGE_MAX - random_float(80)

        #  Current
        current[0] = CURRENT_MAX - random_float(1003)


        for i in range(1, size):
            soc[i] = soc[i - 1] - (soc[0] - SOC_MIN) / size + random_float(100000) * temp * random_int(-1, 1)
            soh[i] = soh[i - 1] - (soh[0] - SOH_MIN) / size + random_float(10000000) * temp * random_int(0, 1)

            voltage[i] = voltage[i - 1] - (voltage[0] - VOLTAGE_MIN) / size + random_float(100000) * temp * random_int(-1, 0)

            if voltage[i] > 5 or voltage[i] < 0:
                print(i)
                print(voltage)
                print("ERROR 4 VOLTAGE")
                sys.exit()

            current[i] = current[i - 1] - (CURRENT_MAX - 0) / size + random_float(10000) * random_int(-1, 0)
            if current[i] > 2 or current[i] < 0:
                print(i)
                print(current[i])
                print("ERROR 4 CURRENT")
                sys.exit()

        # 将每个数组保存到DataFrame
        df = pd.DataFrame({
            'SOC': soc,
            'SOH': soh,
            'Voltage': voltage,
            'Current': current,
            'Temp': np.full(SIZE, temp)  # 将温度列添加到DataFrame中
        })

        all_data.append(df)  # 将每个DataFrame添加到列表中

# 将所有的数据保存到CSV文件中
combined_df = pd.concat(all_data, ignore_index=True)
combined_df.to_csv('all_data.csv', index=False)

print('All data saved successfully.')
```

```python
import numpy as np
import pandas as pd
import sys

# 定义常量和初始值
SOC_MAX = 0.95
SOC_MIN = 0.0

SOH_MAX = 0.99
SOH_MIN = 0.60

VOLTAGE_MAX = 4.2
VOLTAGE_MID = 1.7
VOLTAGE_MIN = 0

CURRENT_MAX = 2

SIZE = 180
SIZE_MAX = 180
SIZE_MIN = 170

TEMP_MAX = 32
TEMP_MIN = 26


def random_int(low, high):
    return np.random.randint(low, high + 1)

def random_float(divisor):
    return np.random.rand() / divisor


all_data_soc = []  # to store data for csv_soc
all_data_soh = []  # to store data for csv_soh

# 循环温度范围
for temp in range(26, 32):

    for num in range(100):
        size = random_int(SIZE_MIN, SIZE_MAX)
        soc = np.zeros(size)
        soh = np.zeros(size)
        voltage = np.zeros(size)
        current = np.zeros(size)

        # SOC
        if random_int(0,1):
            soc[0] = SOC_MAX + random_float(80)
        else:
            soc[0] = SOC_MAX - random_float(80)

        # SOH
        if random_int(0,1):
            soh[0] = SOH_MAX + random_float(80)
        else:
            soh[0] = SOH_MAX - random_float(80)

        # Voltage
        if random_int(0,1):
            voltage[0] = VOLTAGE_MAX + random_float(80)
        else:
            voltage[0] = VOLTAGE_MAX - random_float(80)

        #  Current
        current[0] = CURRENT_MAX - random_float(1003)

        for i in range(1, size):
            soc[i] = soc[i - 1] - (soc[0] - SOC_MIN) / size + random_float(100000) * temp * random_int(-1, 1)
            soh[i] = soh[i - 1] - (soh[0] - SOH_MIN) / size + random_float(10000000) * temp * random_int(0, 1)

            voltage[i] = voltage[i - 1] - (voltage[0] - VOLTAGE_MIN) / size + random_float(100000) * temp * random_int(-1, 0)

            if voltage[i] > 5 or voltage[i] < 0:
                print(i)
                print(voltage)
                print("ERROR 4 VOLTAGE")
                sys.exit()

            current[i] = current[i - 1] - (CURRENT_MAX - 0) / size + random_float(10000) * random_int(-1, 0)
            if current[i] > 2 or current[i] < 0:
                print(i)
                print(current[i])
                print("ERROR 4 CURRENT")
                sys.exit()

        # Calculate time based on voltage values
        time = (VOLTAGE_MAX - voltage + 0.3) / (VOLTAGE_MAX - VOLTAGE_MIN) * (180 - 1) + 13333*random_float(10000) * random_int(0, 1)

        # 将每个数组保存到DataFrame
        df_soc = pd.DataFrame({
            'SOC': soc,
            'Voltage': voltage,
            'Current': current,
            'Temp': np.full(size, temp),  # 将温度列添加到DataFrame中
        })

        df_soh = pd.DataFrame({
            'SOH': soh,
            'Voltage': voltage,
            'Time': time  # 添加时间列
        })

        all_data_soc.append(df_soc)  # 将每个DataFrame添加到列表中
        all_data_soh.append(df_soh)

# 将所有的数据保存到CSV文件中
combined_df_soc = pd.concat(all_data_soc, ignore_index=True)
combined_df_soh = pd.concat(all_data_soh, ignore_index=True)

combined_df_soc.to_csv('csv_soc.csv', index=False)
combined_df_soh.to_csv('csv_soh.csv', index=False)

print('All data saved successfully.')
```
